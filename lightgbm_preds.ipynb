```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import lightgbm as lgbm
%matplotlib inline
```

# LightGBM price prediction model

A model designed simply to predict the price of a product after learning from others.

This is an intermediate step before having an experimental model to optimised margin. This model would provide a measure of how far some actual products' prices is from the modelled solution space. These distances would then be used as price constraints in a NN optimising price.

## Load data

Load a subset of the transaction data and join metadata

```python
# Get the transaction data in
transaction_df = pd.read_csv('data/additional_data/trimmed_transactions.csv')
transaction_df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
print("nrows in data set: {:d}".format(len(transaction_df)))
```

```python
# Load store metadata and join
store_df = pd.read_csv('data/2_store/store.csv', encoding = 'ISO-8859-1')
store_df.set_index('STORE_ID', inplace=True)
```

```python
df = transaction_df.join(store_df, on='STORE_ID')
del store_df, transaction_df
```

```python
# Load product metadata and join
product_df = pd.read_csv('data/2_product/product.csv', encoding = 'ISO-8859-1')
product_df.set_index('PRODUCT_ID', inplace=True)
```

```python
df = df.join(product_df, on='PRODUCT_ID')
del product_df
```

```python
# Load temperature data
temp_df = pd.read_csv('data/additional_data/hk_weather_2016.csv')
temp_df.set_index('Date', inplace=True)
```

```python
df = df.join(temp_df, on='BUSINESS_DATE', how='left')
df.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)
```

```python
# Subsample further at it is still so big!
n_rows = 1000000
df = df.sample(n_rows)
df.info()
```

## Data preparation

Clean the data and encode categoricals

```python
# Effect of location type on the average price in store
# sns.barplot(x=df['TYPE'], y=df['ITEM_AMOUNT'])
```

```python
# Multiple currencies might confuse things! Actually 2, MOK is very similar to HKD (.97). Obviously not cool to keep it long term
# but for now preserve the data volume
assert len(df['CURRENCY_CODE'].unique()) == 2, 'Multiple currencies! Watch out'
```

```python
# Can't learn much from the ID numerical data
df.drop(['STORE_ID', 'PRODUCT_ID', 'MEMBER_ID' ,'CONTACT_ID'], axis=1, inplace=True)
```

```python
# Don't need no closed/closing stores
df = df[df['CLOSING_DATE'].isnull()]
df.drop('CLOSING_DATE', axis=1, inplace=True)
print("nrows in data set after removing closed stores: {:d}".format(len(df)))
```

```python
# Assuming no brand is own brand
df['BRAND_NAME'].fillna('OWN BRAND', inplace=True)
```

```python
# Create the actual label
df['PRICE'] = df['ITEM_AMOUNT'] / df['ITEM_QUANTITY']
df.drop(['ITEM_AMOUNT', 'ITEM_QUANTITY'], inplace=True, axis=1)
```

```python
# Eliminate outlying targets. Use 6 sigma after removing stuff over $1000 - don't want risky predictions at an early model stage
sigma = df['PRICE'].std()
df = df[df['PRICE'] <= 1000]
df = df[(df['PRICE'] <= 3 * sigma) & (df['PRICE'] > 0)]
print("nrows in data set after removing outlying prices: {:d}".format(len(df)))
```

```python
# Look at label distribution. Skewed toward low prices but no biggy for a tree model
sns.distplot(df['PRICE'])
```

```python
# Label encode categorical variables
categorical_vars = df.columns[df.dtypes == 'object'].values
print('Encoded categorical features: \n')
for var in categorical_vars:
    print(var)
    df[var] = LabelEncoder().fit_transform(df[var])
```

```python
# Convert business date into something more meaningful - month
df['BUSINESS_DATE'] = df['BUSINESS_DATE'].astype(str).str.slice(4, 6).astype(int)
```

## Train model

```python
#Separate label
y = df.pop('PRICE')
```

```python
# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=.2, random_state=0)
```

```python
# Create an eval set too
X_train, X_eval, y_train, y_eval = train_test_split(X_train, y_train, test_size=.2)
```

```python
# Create an LGBM data set and set parameters
X_train_lgbm = lgbm.Dataset(data=X_train, label=y_train, categorical_feature=categorical_vars.tolist(), free_raw_data=False)
X_eval_lgbm = lgbm.Dataset(data=X_eval, label=y_eval, categorical_feature=categorical_vars.tolist(), free_raw_data=False)
lgbm_params = {
    'application': 'huber',
    'min_data_in_leaf': 40,
    'boosting_type ': 'dart',
    'learning_rate': 0.1,
    'num_leaves': 31,
    'metric': 'l2'
}
```

```python
# CV for number of boosting rounds
# lgbm.cv(params=lgbm_params, train_set=X_train_lgbm, nfold=5,
#         verbose_eval=10, num_boost_round=200, early_stopping_rounds=50)
```

```python
# Train a regression model
evals_result = {}
reg = lgbm.train(params=lgbm_params, 
                 train_set=X_train_lgbm, 
                 valid_sets=[X_train_lgbm, X_eval_lgbm],
                 valid_names=['train', 'eval'],
                 verbose_eval=10,
                 early_stopping_rounds=20,
                 num_boost_round=200,
                 evals_result=evals_result
                 )
```

```python
fig, ax = plt.subplots(1,1,figsize=[10,5])
ax.plot(evals_result['train']['l2'], label='train')
ax.plot(evals_result['eval']['l2'], label='eval')

ax.set_ylabel('L2 loss')
ax.set_xlabel('Boosting round')
ax.set_title('Training curves')
ax.legend()
ax.grid()
plt.show()
```

```python
# Predict
y_pred = reg.predict(X_test)
y_pred_train= reg.predict(X_train)
```

## Evaluate performance

```python
from sklearn.metrics import r2_score

# Traditional r2. Check train and test for overfitting
r2_train = r2_score(y_train, y_pred_train)
print('Model r2 score train = {:.4f}'.format(r2_train))

r2_test = r2_score(y_test, y_pred)
print('Model r2 score test = {:.4f}'.format(r2_test))
```

```python
# Define a more helpful metric
def predictions_within_threshold(y_test, y_pred, threshold):
    n_preds_within_thresh = sum((y_pred >= threshold*y_test) & (y_pred >= threshold*y_test))
    perc_within_thresh = n_preds_within_thresh / len(y_pred)
    print('{:.2f}% of predictions were within +- {:.1f}% of actual price'.format(100*perc_within_thresh, 100*(1-threshold)))
```

```python
thresholds = [.7, .8, .9, .95, .99]
for thresh in thresholds:
    predictions_within_threshold(y_test, y_pred, thresh)
```

```python
fig, ax = plt.subplots(1, 1, figsize=[9, 7])
ax.scatter(y_test, y_pred, alpha=.2)
ax.set_xlabel('Actual price')
ax.set_ylabel('Predicted price')

ax.set_xlim(0, 500)
ax.set_ylim(0, 500)

plt.show()
```
