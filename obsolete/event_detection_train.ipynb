```python
weather_df = pd.read_csv('data/additional_data/hk_weather_2016.csv')
weather_df = weather_df[weather_df['Date'] >= start_date]
weather_df = weather_df[weather_df['Date'] <= end_date]
```

### 2. Get linear correlations with weather

```python
unique_products_sample = np.random.choice(transaction_df['PRODUCT_ID'].unique(), 2000)
n_products = len(unique_products_sample)
```

```python
scores = []

for i in range(n_products):
    prod = unique_products_sample[i]
    product_df = transaction_df[transaction_df['PRODUCT_ID'] == prod].groupby(by='BUSINESS_DATE')['ITEM_QUANTITY'].sum()

    if len(product_df) > 20:
        weather_ts = pd.Series(weather_df['Mean (deg. C)'].values, index=pd.to_datetime(weather_df['Date'], format='%Y%m%d'))
        product_ts = pd.Series(product_df.values, index=pd.to_datetime(product_df.index.values, format='%Y%m%d'))
        scores.append(weather_ts.corr(product_ts))
    else:
        scores.append(0)
```

```python
scored_products = pd.DataFrame({'products': unique_products_sample,
                               'weather_correlation': scores}).sort_values(by='weather_correlation', ascending=False)
```

Given correlations, cluster products on this

```python
kmeans = KMeans(n_clusters=30, random_state=0).fit(scored_products['weather_correlation'].fillna(0).values.reshape(-1, 1))

cluster_average_corr = []
for cluster in np.unique(kmeans.labels_):
     cluster_average_corr.append(scored_products.loc[kmeans.labels_ == cluster, 'weather_correlation'].mean())
        
cluster_scores = pd.DataFrame({'cluster': np.unique(kmeans.labels_),
                               'average_corr': cluster_average_corr}).sort_values(by='average_corr', ascending=False)
cluster_scores.reset_index(inplace=True,drop=True)

most_correlated_cluster = cluster_scores['cluster'][0]

top_products = scored_products[kmeans.labels_ == most_correlated_cluster]
```

```python
weather_df['Date'] = pd.to_datetime(weather_df['Date'], format='%Y%m%d')
```

Plot cluster with the strongest correlation

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(weather_df['Mean (deg. C)'].min(), weather_df['Mean (deg. C)'].max()), copy=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=[17, 12])

ax.plot(weather_df['Date'], weather_df['Mean (deg. C)'], label=('Temperature'), linewidth=5, color='k')
ax.set_ylabel('Temperature', color='k')
ax.set_xlabel('Time')
ax.tick_params('y', colors='k')

ax2 = ax.twinx()
for prod in top_products['products']:
    prod_data = transaction_df[transaction_df['PRODUCT_ID'] == prod].groupby(by='BUSINESS_DATE')['ITEM_QUANTITY'].sum()
    #ax2.plot(pd.to_datetime(prod_data.index.values, format='%Y%m%d'), prod_data.values, label=prod, linestyle='--')
    ax2.plot(pd.to_datetime(prod_data.index.values, format='%Y%m%d'), scaler.fit_transform(prod_data.values.reshape(-1, 1)),
             label=prod, linestyle='--')
    
ax2.set_ylabel('Sales (normalised to temp scale)')
ax2.legend()
plt.show()
```

## Stage 2: Forecast

Train a simple lightGBM model to get the temperature and sales relationship 

```python
i = 0
for prod in top_products['products']:
    prod_data = transaction_df[transaction_df['PRODUCT_ID'] == prod].groupby(by='BUSINESS_DATE')['ITEM_QUANTITY'].sum()
    weather_ts = pd.DataFrame({'temp': weather_df['Mean (deg. C)'].values}, index=pd.to_datetime(weather_df['Date'], format='%Y%m%d'))
    product_ts = pd.DataFrame({'sales': product_df.values}, index=pd.to_datetime(product_df.index.values, format='%Y%m%d'))
    
    if i == 0:
        temp_vs_sales_df = weather_ts.join(product_ts, how='left').fillna(0)
    else:
        temp_vs_sales_df = pd.concat([temp_vs_sales_df, weather_ts.join(product_ts, how='left').fillna(0)])
        
    i += 1
```

```python
# Feature = temp
# Label = sales
train = lgbm.Dataset(data=temp_vs_sales_df['temp'].values.reshape(-1,1), label=temp_vs_sales_df['sales'].values)
```

```python
params = {'application': 'regression'}
```

```python
reg = lgbm.train(params=params, train_set=train)
```

```python
train_sales_pred = reg.predict(temp_vs_sales_df['temp'].values.reshape(-1,1))
```

```python
from sklearn.linear_model import Ridge, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

temp_lin =Ridge()
temp_lin.fit(temp_vs_sales_df.index.dayofyear.values.reshape(-1,1), temp_vs_sales_df['temp'].values)

sales_lin = Ridge()
sales_lin.fit(temp_vs_sales_df.index.dayofyear.values.reshape(-1,1), temp_vs_sales_df['sales'].values)
```

```python
train_temp_pred = temp_lin.predict(temp_vs_sales_df.index.dayofyear.values.reshape(-1,1))
train_sales_pred = sales_lin.predict(temp_vs_sales_df.index.dayofyear.values.reshape(-1,1))
```
